```python
 from pyspark.sql import SparkSession
from pyspark.sql.functions import lit,when
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
```

```python
def compare_schemas(schema_1,schema_2):
    f1={item.name : item.dataType for item in schema1}
    f2={item.name : item.dataType for item in schema2}
    keys = set(f1.keys()).union(set(f2.keys()))
    discrepancies = []
    for key in keys:
        if key not in f1:
            discrepancies.append(f"column{key} is missing in schema1")
        if key not in f2:
            discrepancies.append(f"column{key} is missing in schema2")
        if key in f1 and key in f2 and f1[key] != f2[key]:
            discrepancies.append(f"Column {key} has different types: {f1[key]} vs {f2[key]}")
    return discrepancies

def withs_col(df1,df2):
    df1_cols= set(df1.columns)
    for col in df1_cols:
        if col not in df2.columns:
            df2=df2.withColumn(col, lit(None).df1.schema[col].dataType)
    return df1,df2
```
